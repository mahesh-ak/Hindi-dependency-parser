```python

words, tags = set([]), set([])
for data in data_hn:
    for form,tag in data:
        tags.add(tag)

for data in train_data_hn:
    for form,tag in data:
        words.add(form)

word2index = {w: i + 1 for i, w in enumerate(list(words))}
tag2index = {t: i + 1 for i, t in enumerate(list(tags))}
word2index['-OOV-']=0
tag2index['UNK']=0
```

```python
test_sentences_X , test_tags_y = [] , []
for s in test_data_hn:
    s_int = [] 
    for w,_ in s:
        try:
            s_int.append(word2index[w]) 
        except KeyError:
            s_int.append(word2index['-OOV-'])
    test_tags_y.append([tag2index[t] for _,t in s])
    test_sentences_X.append(s_int)

```

```python
train_sentences_X , train_tags_y = [] , []
for s in train_data_hn:
    s_int = [] 
    for w,_ in s:
        try:
            s_int.append(word2index[w]) 
        except KeyError:
            s_int.append(word2index['-OOV-'])
    train_tags_y.append([tag2index[t] for _,t in s])
    train_sentences_X.append(s_int)

```

```python
MAX_LENGTH = len(max(train_sentences_X,key=len))

from keras.preprocessing.sequence import pad_sequences
train_sentences_X = pad_sequences(train_sentences_X, maxlen=MAX_LENGTH, padding='post')
train_tags_y = pad_sequences(train_tags_y, maxlen=MAX_LENGTH, padding='post')
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, LSTM, InputLayer, Bidirectional, TimeDistributed, Embedding, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential() 
model.add(InputLayer(input_shape=(MAX_LENGTH, ))) 
model.add(Embedding(len(word2index), 128)) 
model.add(Bidirectional(LSTM(256, return_sequences=True))) 
model.add(TimeDistributed(Dense(len(tag2index)+1))) 
model.add(Activation('softmax')) # we compile the model 

model.compile(loss='categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])
```

```python
model.fit(train_sentences_X, to_categorical(train_tags_y, len(tag2index)+1), batch_size=128, epochs=40, validation_split=0.2)
```

```python
def logits_to_tokens(sequences, index): 
    token_sequences = [] 
    for categorical_sequence in sequences: 
        token_sequence = [] 
        for cate in categorical_sequence: 
            token_sequence.append(index[np.argmax(cate)]) 
        token_sequences.append(token_sequence) 
    return token_sequences
```
