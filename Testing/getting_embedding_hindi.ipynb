```python
import conllu
data_file = open("/home/tushar/Desktop/NLP/Project/UD_Hindi-HDTB-master/hi_hdtb-ud-train.conllu", "r", encoding="utf-8")
list1 = []
for tokenlist in conllu.parse_incr(data_file):
    list1.append(tokenlist)
```

```python
#getting sentences out of dictionary
strlis =  []
for i in range(len(list1)):
    abstr = ""
    for j in range(len(list1[i])):
        abstr += conllu.TokenList(list1[i]).tokens[j]['form']
        abstr += " "
    strlis.append(abstr)
    
print(strlis[10])
```

```python
from bert_embedding import BertEmbedding
sentences = strlis
bert_embeddings = BertEmbedding(model='bert_12_768_12', dataset_name='wiki_multilingual_cased')

```

```python
result = bert_embeddings(sentences)
```

```python
print(type(result))
set1 = {"": []}
```

```python
for i in range(len(result)):
    for j in range(len(result[i][0])):
        set1[result[i][0][j]] = result[i][1][j]
        
print(len(set1))
```
